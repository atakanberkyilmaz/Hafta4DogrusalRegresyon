# Doğrusal Regresyon

```python
#Liste 7.1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Colab\ Notebooks
veriseti = pd.read_csv('DogrusalRegresyon.csv')
veriseti.head()
```

```python
#Liste 7.2
#X = veriseti.iloc[:, 2].values
#y = veriseti.iloc[:, 3].values
X = veriseti['Bas_cevresi(cm^3)'].values
y = veriseti['Beyin_agirligi(gr)'].values
uzunluk = len(X)
X = X.reshape((uzunluk, 1))
```

```python
#Liste 7.3
# Eğitim ve test veri setinin oluşturulması
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Liste 7.4
# Eğitim setinin lineer regresyona göre uydurulması
from sklearn.linear_model import LinearRegression
model_Regresyon = LinearRegression()
model_Regresyon.fit(X_train, y_train)
```

```python
#Liste 7.5
# Test seti sonuçlarının tahmin edilmesi
y_pred = model_Regresyon.predict(X_test)
```

```python
#Liste 7.6
# Eğitim seti sonuçlarının grafiğinin çizilmesi
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model_Regresyon.predict(X_train), color = 'blue')
plt.title('Başın çevre uzunluğu ve beyin ağırlığı (eğitim veri seti)')
plt.xlabel('Başın çevre uzunluğu (cm^3)')
plt.ylabel('Beyin ağırlığı (gram)')
plt.show()

# Test seti sonuçlarının grafiğinin çizilmesi
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, model_Regresyon.predict(X_train), color = 'blue')
plt.title('Başın çevre uzunluğu ve beyin ağırlığı (test veri seti)')
plt.xlabel('Başın çevre uzunluğu (cm^3)')
plt.ylabel('Beyin ağırlığı (gram)')
plt.show()
```

```python
#Liste 7.7
# Regresyon denklemi katsayıları
print('Eğim(Q1):', model_Regresyon.coef_) # x in önündeki katsayıyı verir
print('Kesen(Q0):', model_Regresyon.intercept_) # + b değerini verir
#Regresyon denklemi
print("y=%0.2f"%model_Regresyon.coef_+"x+%0.2f"%model_Regresyon.intercept_)
```

```python
#Liste 7.8
# Test veri seti performansı
from sklearn.metrics import explained_variance_score , mean_absolute_error , mean_squared_error
from sklearn.metrics import median_absolute_error , r2_score # hata değerleri ve rkare değeri hesaplanır
print("R-Kare: ",r2_score(y_test, y_pred))
print("MAE: ",mean_absolute_error(y_test, y_pred))
print("MSE: ",mean_squared_error(y_test, y_pred))
print("MedAE: ",median_absolute_error(y_test, y_pred))
```
